```python
import pandas as pd
import re

# Load the dataset
df = pd.read_csv('tamil_news_train.csv')

# Extract month and year using regex
def extract_month_year(news):
    match = re.search(r'(\d{4}/\w{3})', news)
    return match.group(1) if match else None

df['Month_Year'] = df['NewsInEnglish'].apply(extract_month_year)
```
